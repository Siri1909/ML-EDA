```python
#3.VISULISATION - seabron - for age
import seaborn as sns
sns.distplot(df['age'])
```

```python
#This count plot will tell us number of ages
import matplotlib.pyplot as plt
f=plt.figure()
f.set_figwidth(12)
f.set_figheight(5)
plt.plot()
sns.countplot(x = 'age',data = df)
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
df.groupby('age').size()
```

```python
df['age'].value_counts()
```

```python
import matplotlib.pyplot as plt
f=plt.figure()
f.set_figwidth(15)
f.set_figheight(5)
plt.plot()
df['age'].value_counts().plot(kind = 'bar')
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
import matplotlib.pyplot as plt
f=plt.figure()
f.set_figwidth(12)
f.set_figheight(5)
plt.plot()
sns.barplot(x = df['sex'],y = df['age'])
plt.xticks(rotation = 'horizontal')
plt.show()
```

```python
np.max(df['age'])
```
